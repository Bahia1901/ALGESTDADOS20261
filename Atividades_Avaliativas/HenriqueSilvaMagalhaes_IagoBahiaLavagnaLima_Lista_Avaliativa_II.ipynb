```python
#1 Uma empresa de médio porte deseja realizar um levantamento sobre a remuneração de seus funcionários. Desenvolva um programa que leia os salários de 10 funcionários, um por vez, e, ao final, informe a média salarial da empresa, o maior salário registrado e quantos funcionários recebem acima de R$ 3.000. Os valores informados são válidos e positivos.
soma = 0
acima_30002 = 0

for i in range(1, 11):

    salario = float(input(f"Digite o salário do {i}° funcionário: "))

    soma += salario

    if i == 1:
        maior = salario

    if salario > maior:
        maior = salario

    if salario > 3000:
        acima_3000 += 1

media = soma / 10

print(f"Média salarial: R$ {media:.2f}")
print(f"Maior salário: R$ {maior:.2f}")
print(f"Funcionários que recebem acima de R$ 3000: {acima_3000}")
```

```python
#2 Em uma instituição de ensino, um professor deseja analisar o desempenho de uma turma sem saber previamente a quantidade de alunos. Desenvolva um programa que leia as notas dos alunos até que seja digitado o valor -1 (fim da entrada). Ao final, calcule a média da turma, informe quantos alunos foram aprovados (nota ≥ 7) e quantos foram reprovados.
soma = 0
quantidade = 0
aprovados = 0
reprovados = 0

nota = float(input("Digite a nota do aluno (-1 para parar): "))

while nota != -1:

    soma += nota
    quantidade += 1

    if nota >= 7:
        aprovados += 1

    else:
        reprovados += 1

    nota = float(input("Digite a nota do aluno (-1 para parar): "))

media = soma / quantidade

print(f"Média da turma: {media:.2f}")
print(f"Quantidade de aprovados: {aprovados}")
print(f"Quantidade de reprovados: {reprovados}")
```

```python
#3 Um sistema bancário simples precisa simular operações de saque em uma conta corrente. O programa deve iniciar com um saldo informado pelo usuário e permitir saques sucessivos até que seja digitado 0. A cada tentativa, verifique se há saldo suficiente, atualize o saldo quando possível e informe ao usuário quando o valor exceder o saldo disponível

saldo = float(input("Digite o saldo da conta: "))

saque = float(input("Digite o valor do saque (0 para sair): "))

while saque != 0:

    if saque <= saldo:

        saldo -= saque

        print(f"Saque realizado com sucesso")
        print(f"Saldo atual: R$ {saldo:.2f}")

    else:
        print("Saldo insuficiente")

    saque = float(input("Digite o valor do saque (0 para sair): "))

print(f"Saldo final: R$ {saldo:.2f}")
```

```python
#5 Um sistema de segurança controla o acesso a um ambiente restrito. O programa deve solicitar uma senha numérica e permitir até três tentativas. Caso a senha correta (2024) seja informada dentro desse limite, o acesso deve ser liberado. Caso contrário, após três tentativas incorretas, o sistema deve bloquear o acesso e informar essa condição ao usuário.

senha_correta = 2024
tentativas = 0

while tentativas < 3:

    senha = int(input("Digite a senha: "))

    if senha == senha_correta:
        print("Acesso liberado")
        break

    else:
        tentativas += 1
        print("Senha incorreta")

if tentativas == 3:
    print("Acesso bloqueado")
```

```python
#28 Um sistema deve analisar números inteiros informados pelo usuário até que seja digitado -1. Ao final, informe quantos números são primos e calcule a soma dos números que não são primos.
primos = 0
soma_nao_primos = 0

numero = int(input("Digite um número (-1 para parar): "))

while numero != -1:

    divisores = 0

    for i in range(1, numero + 1):

        if numero % i == 0:
            divisores += 1

    if divisores == 2:
        primos += 1

    else:
        soma_nao_primos += numero

    numero = int(input("Digite um número (-1 para parar): "))

print(f"Quantidade de números primos: {primos}")
print(f"Soma dos números não primos: {soma_nao_primos}")
```
